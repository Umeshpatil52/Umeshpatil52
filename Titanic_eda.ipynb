```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Titanic-Dataset.csv")  #make sure the file is in same folder

df.head(20)
```

```python
df.info()
```

```python
df.describe()                     #only used for int and float dtypes
```

```python
df.isnull().sum()
```

```python
#Explore Categorical Columns
df['Sex'].value_counts().plot(kind='bar')
```

```python
df['Embarked'].value_counts().plot(kind='bar')
```

```python
#explore numerical columns
sns.histplot(df['Age'].dropna(), kde=True)
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
```

```python
#correlation analysis
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
#Insights
# 1. negative correlation between Fare and Pclass as the Pclass increases numerically (1  3), Fare actually decreases 
# 2. boxplot helps in finding outliers easily
# 3. KDE (Kernel Density Estimate) curve estimates the probability density function of the data.
# 4. Categorical Columns help in finding basic info like embarked column showed most of passengers boarded from Southampton (England)
# 5. describe function is only used for int and float dtypes
# 6. make sure the csv file is in same folder 
# 7. I know these are easy insights but I found it helpful while analysing this dataset!
```
